Also, we have to construct the label variable: Label variable is 1 if project doesn't get fully funded within the 60 days from date posted, 0 otherwise

```python
data['days_financed'] = pd.to_datetime(data['datefullyfunded']) - pd.to_datetime(data['date_posted'])
data['label'] = data['days_financed'].dt.days.apply(lambda x: 0 if x <= 60 else 1)
```

##  Step 3. Pre-process the data

We will make this We drop the attributes that are not necessary for prediction. These are 'projectid', 'teacher_acctid', 'schoolid','school_ncesid', which are identifiers. 
Also, although very important, we will leave behind geographic location attributes:          'school_longitude', 'school_district', 'school_county', 'school_city' and 'school_state', as well as the  attributes used to construct the label:  'datefullyfunded', 'days_financed'

The missing values of the 'primary_focus_area','resource_type', 'grade_level', 'students_reached' will be filled after the temporal validation.  We won't do so with secondary_focus_area because there are plenty of missings and would bias our models.


```python
drop_lst = ['projectid', 'teacher_acctid', 'schoolid',
            'school_ncesid', 'school_latitude', 
            'school_longitude', 'school_district', 
            'school_county', 'secondary_focus_subject',
            'secondary_focus_area','primary_focus_subject']
unnecesary_lst = ['datefullyfunded', 'days_financed', 'school_city', 'school_state']
pl.drop_features(data, drop_lst)
pl.drop_features(data, unnecesary_lst)


lst_to_fill = ['primary_focus_area',
               'resource_type', 'grade_level', 'students_reached']
pl.fill_missing(data,lst_to_fill,form = False)

lst_dummies = ['school_metro', 
               'primary_focus_area', 'resource_type', 'poverty_level',
              'grade_level']
data = pl.to_dummies(data,lst_dummies)

pl.drop_features(data, lst_dummies)
```

## Step 4. Generate features/predictors
Features will be created in the loop, once the temporal validation windows have been stablished
Do additional transformations

```python
d_t = pd.get_dummies(data['teacher_prefix'])
d_t_lst = list(d_t.columns)
d_t['female_teacher'] = d_t['Dr.'] + d_t['Ms.'] + d_t['Mrs.'] 
d_t['male_teacher'] = d_t['Mr.']
pl.drop_features(d_t, d_t_lst)
data.join(d_t)

final_drop = ['teacher_prefix']
pl.drop_features(data, final_drop)
```

An additional function that is also necesary is the boolean conversor from string to 1/0. Some of the variables are in "t", "f" and we want them to be integers 1/0

```python
tf_dic = {'t': 1, 'f':0}
data = data.replace({'school_charter': tf_dic, 'school_magnet': tf_dic, 'eligible_double_your_impact_match' : tf_dic})
```

## Step 5. Build classifier

First we define the temporal window we want to split the data on. For this exercise, we take windows of 6 months, using 6 months for training and the following 6 months for testing, and so on. We have 3 validation sets. 

```python
temp_lst = ep.temp_val('2012-01-01','2013-12-31',6,6,60)
temp_lst
```

The following runs a menu of models that can be chosen between the following: Bagging, Random Forests, Decision Trees, k Nearest neighbors, gradient boosting, support vector machines. 

### 5.1  For this case, we will re-read the data so we don't use the already pre-processed data we used to evaluate the trends. 

```python
data = pl.read_data('/Users/csolisu/Downloads/projects_2012_2013.csv')
data['days_financed'] = pd.to_datetime(data['datefullyfunded']) - pd.to_datetime(data['date_posted'])
data['label'] = data['days_financed'].dt.days.apply(lambda x: 0 if x <= 60 else 1)
d_t = pd.get_dummies(data['teacher_prefix'])
d_t_lst = list(d_t.columns)
d_t['female_teacher'] = d_t['Dr.'] + d_t['Ms.'] + d_t['Mrs.'] 
d_t['male_teacher'] = d_t['Mr.']
pl.drop_features(d_t, d_t_lst)
data.join(d_t)
final_drop = ['teacher_prefix']
pl.drop_features(data, final_drop)
tf_dic = {'t': 1, 'f':0}
data = data.replace({'school_charter': tf_dic, 'school_magnet': tf_dic, 'eligible_double_your_impact_match' : tf_dic})
temp_lst = ep.temp_val('2012-01-01','2013-12-31',6,6,60)
```

```python
clfs, grid = ep.define_clfs_params('small')
models_to_run=['BAG','RF','DT','KNN','LR','GB','ET']
metric_lst =[ep.precision_at_k,ep.recall_at_k,ep.f1_at_k]
k_lst = [1.0,2.0,5.0,10.0,20.0,30.0,40.0,50.0]
results_df_long = ep.clf_loop(models_to_run,clfs,grid,data,'date_posted','label',temp_lst,metric_lst,k_lst,10)
```

```python
results_df_long
```

```python
df_1 = ep.rank_classifiers(results_df_long,'train_end_date','roc_auc')
```

```python
df_1.sort_values(by='2013-05-02 00:00:00', ascending= True, inplace= True)
df_1
```

We graph the average evaluation metrics for the six types of models in the previous section. The chosen threshold is at 5% because we want to get the 5% of projects that are at risk of not getting fully funded within the first 60 days.

```python
graph = results_df_long.groupby(['model_type'], sort=False)['auc-roc', 'p_at_5', 'r_at_5', 'f1_at_5'].mean()
grph = graph.plot(kind='barh', figsize=[12,6], 
                     alpha=.4, legend=True).get_figure()
grph.savefig('mean_performance.pdf')
```

We graph the maximum evaluation metrics for the six types of models in the previous section. We see that the model that gets better performance is the Gradient Boosting one.  The chosen threshold is at 5% because we want to get the 5% of projects that are at risk of not getting fully funded within the first 60 days.

```python
graph = results_df_long.groupby(['model_type'], sort=False)['auc-roc', 'p_at_5', 'r_at_5', 'f1_at_5'].max()
grph =graph.plot(kind='barh', figsize=[12,6], 
                     alpha=.4, legend=True).get_figure()
grph.savefig('max_performance.pdf')
```

The best model of them all is chosen with the auc-roc criteria. We see that this model has an auc-roc of 0.6620 for the first temporal window, 0.682 for the second temporal window (an improvement), and 0.674 for the third temporal window. We should determine if this is actually the good model given that the performance seems to be worsening throughout time. 

We observe that the precision achieved by this model at any of the thresholds won't be satisfying.

We observe that the predicted number of values who won't get fully funded is very small. Even though this was the best predicted model, we should refine the features selection and reduce them to the most important ones.
